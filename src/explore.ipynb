# Random Forest Project Tutorial

Predicting diabetes

```python
import pandas as pd

#TOTAL DATA
#Reading the total_data_sel dataset
total_data_train_sel = pd.read_csv("/workspaces/machine-learning-random-forest-algorithm/data/processed/total_data_train_sel.csv")
total_data_test_sel = pd.read_csv("/workspaces/machine-learning-random-forest-algorithm/data/processed/total_data_test_sel.csv")
total_data_sel = pd.concat([total_data_train_sel, total_data_test_sel]).reset_index(drop = True)


X_train_sel = total_data_train_sel.drop(["Outcome"], axis = 1)
X_test_sel = total_data_test_sel.drop(["Outcome"], axis = 1)
y_train = total_data_train_sel["Outcome"]
y_test = total_data_test_sel["Outcome"]
```

```python
#Initialization and training of the model

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state = 8)
model.fit(X_train_sel, y_train)
```

```python

import matplotlib.pyplot as plt
from sklearn import tree

fig, axis = plt.subplots(2, 2, figsize = (15, 15))

# We show the first 4 trees out of the 100 generated (default)
tree.plot_tree(model.estimators_[0], ax = axis[0, 0], feature_names = list(X_train_sel.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model.estimators_[1], ax = axis[0, 1], feature_names = list(X_train_sel.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model.estimators_[2], ax = axis[1, 0], feature_names = list(X_train_sel.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model.estimators_[3], ax = axis[1, 1], feature_names = list(X_train_sel.columns), class_names = ["0", "1", "2"], filled = True)

plt.show()
```

```python

#Model prediction

y_pred = model.predict(X_test_sel)
y_pred
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

The datasets with the best performance are total_data_norm_sel and total_data_scal_sel, so we'll take total_data_norm_sel for the optimization. 

```python
#Model optimization

#Grid search
from sklearn.model_selection import GridSearchCV

# We define the parameters that we want to adjust by hand
hyperparams = {
    "n_estimators": [50, 80, 100],
    "criterion": ["gini", "entropy", "log_loss"],
    "max_depth": [5, 10, 15, None],
    "min_samples_split": [2,6,12],
    "min_samples_leaf": [1,2,4]
    }
```

```python
# We initialize the grid
grid = GridSearchCV(model, hyperparams, scoring = "accuracy", cv = 10)
grid
```

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

grid.fit(X_train_norm_sel, y_train)

print(f"Best hyperparameters: {grid.best_params_}")
```

```python

model_grid = RandomForestClassifier(criterion="gini", max_depth=5, min_samples_leaf=2, min_samples_split=6, n_estimators=100, random_state=8)
model_grid.fit(X_train_sel, y_train)
y_pred = model_grid.predict(X_test_sel)

grid_accuracy = accuracy_score(y_test, y_pred)
grid_accuracy


```

```python
from pickle import dump

dump(model_grid, open("../models/RandomForestClassifier_estimators-100_criterion-gini_minsamplesleaf-2_minsamplessplit-6_8.sav", "wb"))
```
